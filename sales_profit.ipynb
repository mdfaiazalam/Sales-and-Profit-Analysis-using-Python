```python
region_market_sales_profit = df.groupby(['Region', 'Market'], observed = False)[['Sales', 'Profit']].sum()
top_region_sales = region_market_sales_profit.sort_values(by = 'Sales', ascending = False)
top_market_profit = region_market_sales_profit.sort_values(by = 'Profit', ascending = False)
region_market_sales_profit
```

```python
top_region_sales[['Sales']].head()
top_market_profit[['Profit']].head()
```

```python
top_region_sales = region_market_sales_profit.sort_values(by = 'Sales', ascending = False).head()
top_market_profit= region_market_sales_profit.sort_values(by = 'Profit', ascending = False).head()

fig, ax = plt.subplots(1,2,figsize = (14,7))

#sales by market
top_region_sales[['Sales', 'Profit']].plot(kind = 'bar', stacked = True, ax=ax[0], color = ['red', 'blue'])
ax[0].set_title('Top regions by sales and profit', fontsize=14, color='darkblue')
ax[0].set_ylabel('Amount ($)', fontsize=12)
ax[0].set_xlabel('Market', fontsize = 12)

#profit by market
top_market_profit[['Sales', 'Profit']].plot(kind = 'bar', stacked = True, ax = ax[1], color = ['red', 'blue'])
ax[1].set_title('Top market by Sales and Profit', fontsize = 14, color = 'darkblue')
ax[1].set_ylabel('Amount ($)', fontsize = 12)
ax[1].set_xlabel('Market', fontsize = 12)

plt.legend()
plt.tight_layout()

plt.savefig(r'C:\Users\hp\Downloads\project DA\Salees and Profit Python Project\top_region_market_by_sales_profit', dpi = 300, bbox_inches = 'tight')
```

```python
#Cities and states are more profitable

top_profit_city = df.groupby(['City', 'State'])['Profit'].sum().reset_index()
top_profit_state_city = top_profit_city.sort_values(by = 'Profit', ascending = False).head()
top_profit_state_city
```

```python
top_profit_state_city = top_profit_city.sort_values(by = 'Profit', ascending = False).head()

fig, ax = plt.subplots(figsize = (10,6))
top_profit_state_city.plot(kind = 'bar', x = 'City', y = 'Profit', ax = ax, color = 'blue')
ax.set_title('Top city by Profit', color = 'red', fontweight = 'bold', fontsize = 20)
ax.set_ylabel('Amount ($)', fontsize = 13, fontweight = 'bold')
ax.set_xlabel('State', fontweight = 'bold')

plt.xticks(rotation = 45, ha='right')

plt.tight_layout()

plt.savefig(r'C:\Users\hp\Downloads\project DA\Salees and Profit Python Project\most_profitabel_state_city.png', dpi = 300, bbox_inches = 'tight')
```
